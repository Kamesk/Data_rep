```python
Recency = data.groupby(by='Member_number')['Date'].max().reset_index()
Recency.columns = ['memberID', 'LastDate']
last_date_dataset = Recency['LastDate'].max()
last_date_dataset
Recency['Recency'] = Recency['LastDate'].apply(lambda x: (last_date_dataset - x).days)
Recency.head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('groceries_dataset.csv')

# Convert itemDescription column to numeric using TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(data['itemDescription'])

# Perform k-means clustering
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(X)

# Reduce the dimensionality of X using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())

# Add cluster labels to the dataset
data['Cluster'] = kmeans.labels_

# Display cluster information
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('Clusters')
plt.show()
```

```python
# Convert itemDescription column to numeric using TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(data['itemDescription'])

# Predict cluster labels for the new dataset
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(X)
data['Cluster'] = kmeans.predict(X)

# Display the new dataset with predicted cluster labels
print(data)
```
